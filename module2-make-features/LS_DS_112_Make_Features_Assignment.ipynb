```python
# Fixing a single value
int_rate= ' 12.98%'
```

```python
#Writing a function that can be used for any string that is passed 
def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
#Testing out the function
int_rate_to_float(int_rate)
```

```python
#Checking data type
type(int_rate_to_float(int_rate))
```

```python
#Apply the solution to every cell in the column

new_list=[]
for cell in df['int_rate']:
  new_list.append(int_rate_to_float(cell))

new_list[0:10]
  
```

```python
df['int_rate_cleaned']=pd.Series(new_list)
df.head()
```

```python
# Type of data for the new column
type(df['int_rate_cleaned'])
```

```python
# Create some examples that represnt the cases we want to clean up

examples = ['owner', 'Supervisor', 'Project Manager', np.NaN]
```

```python
def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"

for title in examples:
  print(clean_title(title))
```

```python
# Using List Comprehension
[clean_title(example) for example in examples]
```

```python
#Applying the function to the column
df['emp_title']=df['emp_title'].apply(clean_title)
df.head()
```
