## Writing a reusable function

```python
#Writing a function that can be used for any string that is passed 
def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
#Testing out the function
print(int_rate_to_float(int_rate))
print(int_rate_to_float('23%'))
```

```python
#Checking data type
type(int_rate_to_float(int_rate))
```

## Using .apply method

```python
#Apply the solution to every cell in the column

new_list=[]

for cell in df['int_rate']:
  new_list.append(int_rate_to_float(cell))

new_list[:10]
  
```

```python
df['int_rate_cleaned']=pd.Series(new_list)
df.head()
```

```python
# Type of data for the new column
type(df['int_rate_cleaned'])
```

```python
# Data types for the last 5 columns
df.dtypes[-5:]
```

##Improving this code with .apply


```python
# Using .Apply
df['int_rate_cleaned_from_apply']=df['int_rate'].apply(int_rate_to_float)
df.head()
```
